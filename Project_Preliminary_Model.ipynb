```python
import tellurium as te
SI_model = '''
    //Equations
    E1: T -> I1 ; beta * T * V ; //Infection rate
    E2: I1 -> I2 ; k*I1 ; //Infection rate
    E3: -> V ; p * I2 ; //Virus Production
    E4: V -> ; c * V ; //Viral Death Rate
    E5: I2 -> ; Gs * I2 // (Virus-Producing) Infected Cell Death Rate
    
    //Parameters
    beta = 10E-6 ;
    k = 4.0 ;
    p = 10E-1 ;
    c = 1
    Gd = 1
    Kg = 10E2
    
    //Inputs  
    Gs = Gd / (Kg + I2)
    T0 = 10E7 ;
    T = T0 ;
    I1 = 75 ;
'''
m = te.loada(SI_model)
s = m.simulate(0,2.5,1000, ['time', '[T]','[I1]','[I2]'])
m.plot(title = "Cell Population", xtitle="time (days)", ytitle="variable",linewidth=2.5)
m = te.loada(SI_model)
s = m.simulate(0,2.5,1000, ['time', '[V]'])
m.plot(title = "Viral Load", xtitle="time (days)", ytitle="variable",linewidth=2.5, yscale = 'log', ylim = [10, 10E7])
```

```python
b = [10E-6, 10E-5, 10E-4, 10E-3, 10E-2, 10E-1]
for a in range(len(b)):
    m.resetAll()
    m.beta = b[a]
    s = m.simulate(0,2.5,100, ['time', '[T]','[I1]','[I2]'])
    m.plot(title = "Cell Population", xtitle="time (days)", ytitle="variable",linewidth=2.5)
```
